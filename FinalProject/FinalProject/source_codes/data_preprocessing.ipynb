#### üîé Ta th·∫•y c√≥ ƒë√¥i ch·ªó sai ch√≠nh t·∫£ (khi c√≥ d·∫•u khi kh√¥ng c√≥ d·∫•u) trong t√™n th√†nh ph·ªë (City) v√† t·ªânh th√†nh (State) c·ªßa Vi·ªát Nam. Ta c·∫ßn ƒëi·ªÅu ch·ªânh l·∫°i cho th·ªëng nh·∫•t

```python
def convert_city(city_name):
    if city_name == "Ho Chi Minh City":
        return "TP.HCM"
    elif city_name == "Hanoi":
        return "H√† N·ªôi"
    elif city_name == "Da Nang":
        return "ƒê√† N·∫µng"
    elif city_name == "Can Tho":
        return "C·∫ßn Th∆°"
    elif city_name == "Rach Gia":
        return "R·∫°ch Gi√°"
    else:
        return city_name
    
def convert_state(state_name):
    if state_name == "Ho Ch√≠ Minh City":
        return "TP.HCM"
    elif state_name == "Th·ªß D√¥ H√† N·ªôi":
        return "H√† N·ªôi"
    elif state_name == "D√† Nang":
        return "ƒê√† N·∫µng"
    elif state_name == "T·ªânh C·∫ßn Th∆°":
        return "C·∫ßn Th∆°"
    else:
        return state_name
    
df_order["City"] = df_order["City"].apply(convert_city)
df_order["State"] = df_order["State"].apply(convert_state)
```

### üõí 8. M·ªói c·ªôt hi·ªán ƒëang c√≥ ki·ªÉu d·ªØ li·ªáu g√¨? C√≥ c·ªôt n√†o c√≥ ki·ªÉu d·ªØ li·ªáu ch∆∞a ph√π h·ª£p ƒë·ªÉ c√≥ th·ªÉ x·ª≠ l√Ω ti·∫øp kh√¥ng?

```python
col_dtypes_order = pd.Series(df_order.dtypes)
col_dtypes_order
```

```python
col_dtypes_return = pd.Series(df_return.dtypes)
col_dtypes_return
```

#### üîé T√¨m hi·ªÉu v·ªÅ c√°c c·ªôt c√≥ ki·ªÉu object trong `Orders` sheet

```python
def object_dtype(s):
    dtypes = set()
    s.apply(lambda x : dtypes.add(type(x)))
    return dtypes
```

```python
list_object_order = list(col_dtypes_order.loc[col_dtypes_order.values == object].index)
list_type_order = []

for i in list_object_order:
    list_type_order.insert(len(list_type_order), object_dtype(df_order[i]))
    
type_object_order = pd.Series(list_type_order, list_object_order)
print(type_object_order)
```

#### üîé T√¨m hi·ªÉu v·ªÅ c√°c c·ªôt c√≥ ki·ªÉu object trong `Returns` sheet

```python
list_object_return = list(col_dtypes_return.loc[col_dtypes_return.values == object].index)
list_type_return = []

for i in list_object_return:
    list_type_return.insert(len(list_type_return), object_dtype(df_return[i]))
    
type_object_return = pd.Series(list_type_return, list_object_return)
print(type_object_return)
```

##### üí° **Ta th·∫•y r·∫±ng**: n√™n ƒë∆∞a t·∫•t c·∫£ c√°c c·ªôt tr√™n v·ªÅ `string` ƒë·ªÉ ti·∫øp t·ª•c kh√°m ph√° th√™m c√°c c·ªôt n√†y.

<h2> üìÇ QUAY L·∫†I B∆Ø·ªöC KH√ÅM PH√Å D·ªÆ LI·ªÜU </h2>

Trong ph·∫ßn n√†y ta t√¨m hi·ªÉu s·ª± ph√¢n b·ªë c√°c gi√° tr·ªã ·ªü c√°c c·ªôt `categorical` v√† `numeric` c·ªßa `Orders` sheet

### üõí 9. V·ªõi m·ªói c·ªôt c√≥ ki·ªÉu d·ªØ li·ªáu d·∫°ng categorical, ta t√¨m hi·ªÉu s·ª± ph√¢n b·ªë c√°c gi√° tr·ªã

üîé C√°c c·ªôt trong `list_object_order` thu·ªôc nh√≥m `Categorical`.

V·ªõi m·ªói c·ªôt `categorical`, ta th·ª±c hi·ªán t√≠nh % gi√° tr·ªã thi·∫øu (t·ª´ 0 - 100), s·ªë l∆∞·ª£ng gi√° tr·ªã kh√°c nhau (kh√¥ng x√©t gi√° tr·ªã thi·∫øu), list/array c√°c gi√° tr·ªã kh√°c nhau (kh√¥ng x√©t gi√° tr·ªã thi·∫øu) v√† l∆∞u k·∫øt qu·∫£ v√†o `cate_col_df`. Dataframe n√†y c√≥ 3 c·ªôt l√† `"missing_ratio", "num_diff_vals", "diff_vals"`; v√† c√≥ c√°c d√≤ng l√† c√°c c·ªôt d·ªØ li·ªáu thu·ªôc nh√≥m categorical.

```python
col_cate_info = ["missing_ratio", "num_diff_vals", "diff_vals"]
cate_col_df = pd.DataFrame(columns = col_cate_info, index = list_object_order)

def cate_values(col):
    missing_ratios = df_order[col].isna().sum() / len(df_order) * 100
    num_diff = df_order[col].dropna().nunique()
    diff_value = list(df_order[col].dropna().unique())

    return missing_ratios, num_diff, diff_value

for col in list_object_order:
    cate_col_df.loc[col] = cate_values(col)

cate_col_df
```

### üõí 10. V·ªõi m·ªói c·ªôt c√≥ ki·ªÉu d·ªØ li·ªáu d·∫°ng numeric, ta t√¨m hi·ªÉu s·ª± ph√¢n b·ªë c√°c gi√° tr·ªã

üîé Ta x√©t c√°c c·ªôt `Sales, Quantity, Profit, Shipping Cost` thu·ªôc nh√≥m `Numeric`.

V·ªõi m·ªói c·ªôt `numeric`, ta s·∫Ω t√≠nh t·ªâ l·ªá % gi√° tr·ªã thi·∫øu (t·ª´ 0 ƒë·∫øn 100), min, max. Sau ƒë√≥ l∆∞u k·∫øt qu·∫£ v√†o dataframe `nume_col_profiles_df`. Dataframe n√†y c√≥ 6 c·ªôt l√† `"missing_ratio", "min", "lower_quartile", "median", "upper_quartile", "max"` v√† c√≥ c√°c d√≤ng l√† c√°c c·ªôt d·ªØ li·ªáu thu·ªôc nh√≥m numeric.

```python
col_numeric = ["Sales", "Quantity", "Profit", "Shipping Cost"]
col_num_info = ["missing_ratio", "min", "lower_quartile", "median", "upper_quartile", "max"]
num_col_df = pd.DataFrame(columns = col_num_info, index = col_numeric)

def num_values(col):
    missing_ratios = (df_order[col].isna().sum() / len(df_order) * 100)
    min_value = df_order[col].min()
    median = df_order[col].median()
    lower_quartile = df_order[col].quantile(q = 0.25, interpolation = 'lower')
    upper_quartile = df_order[col].quantile(q = 0.75, interpolation = 'higher')
    max_value = df_order[col].max()

    return missing_ratios, min_value, lower_quartile, median, upper_quartile, max_value

for col in col_numeric:
    num_col_df.loc[col] = num_values(col)

num_col_df
```
